```python
%%writefile GradingSystem.py

from student import *
import time
import copy


def checktime(func):
    def new_func(*args,**kwargs):
        start_time = time.time()
        result = func(*args,**kwargs)
        print("\n실행시간은:",(time.time()-start_time))
        return result
    return new_func


class GradingSystem:
    ig = 0
    language = list()
    english = list()
    math = list()
    temp_list = list()
    
    def __init__(self):
        GradingSystem.ig = 0
        self._student_list = list()
    
    @property
    def student_list(self):
        return self._student_list
        
    def register_student(self, x):
        GradingSystem.ig += 1
        dic = dict()
        dic["번호"] = x.index
        dic["이름"] = x.name
        dic["주민번호"] = x.num
        dic["국어"] = x.a
        dic["영어"] = x.b
        dic["수학"] = x.c
        dic["총점"] = x.a + x.b + x.c
        dic["등수"] = 0
        self.student_list.append(dic)
        
        
    def process(self):
        self.student_list.sort(key = lambda x : x["총점"], reverse = True)
        #language = list()
        GradingSystem.language = sorted(self.student_list,key = lambda x : x["국어"], reverse = True)
        #english = list()
        GradingSystem.english = sorted(self.student_list,key = lambda x : x["영어"], reverse = True)
        #math = list()
        GradingSystem.math = sorted(self.student_list,key = lambda x : x["수학"], reverse = True)
        
        temp_score = self.student_list[0]["총점"]
        self.student_list[0]["등수"] = 1
        for k in range(1, GradingSystem.ig):
            if(temp_score == self.student_list[k]["총점"]):
                self.student_list[k]["등수"] = self.student_list[k-1]["등수"]
            else:
                self.student_list[k]["등수"] = self.student_list[k-1]["등수"] + 1
            temp_score = self.student_list[k]["총점"]
        
    def print_students(self):
        print("\n전체 학생 명단입니다. ")
        for x in range(0, GradingSystem.ig):
                print("등수: " + str(self.student_list[x]["등수"]) + " 총점: " + str(self.student_list[x]["총점"]) + " 이름: " + self.student_list[x]["이름"] + " 주민번호: " + self.student_list[x]["주민번호"])
    
    def print_class_info(self):
        print("\n---------------class info---------------")
        a_avg = 0.5 # 국어 평균
        index_a = 0 #국어 최고점 학생 index
        b_avg = 0.5 # 영어 평균
        index_b = 0
        c_avg = 0.5 # 수학 평균
        index_c = 0
        best_each = 0 #각 과목당 최고점과 해당 학생 명단
        
        for i in range(0,GradingSystem.ig):
            a_avg += self.student_list[i]["국어"]
            b_avg += self.student_list[i]["영어"]
            c_avg += self.student_list[i]["수학"]
            
        x = 0
        y = 0
        z = 0
        q = 0
        r = 0
        s = 0
        #과목별 최고점 구하기
        for p in range(0, GradingSystem.ig):
            if(q < self.student_list[p]["국어"]):
                q = self.student_list[p]["국어"]
            if(r < self.student_list[p]["영어"]):
                r = self.student_list[p]["영어"]
            if(s < self.student_list[p]["수학"]):
                s = self.student_list[s]["수학"]
        
        print("국어 총점 : %.2f 국어 평균 : %.2f" %(a_avg, a_avg/5) )
        print("영어 총점 : %.2f 영어 평균 : %.2f" %(b_avg, b_avg/5) )
        print("수학 총점 : %.2f 수학 평균 : %.2f" %(c_avg, c_avg/5) )
    
        print("\n---------------과목별 최고점---------------")
        print("국어 최고점 : %d" %(q))
        for i in range(0, GradingSystem.ig):
            if(self.student_list[i]["국어"] == q):
                print("이름: " + self.student_list[i]["이름"] + " 주민번호: " + self.student_list[i]["주민번호"])
 
        print("영어 최고점 : %d" %(r))
        for i in range(0, GradingSystem.ig):
            if(self.student_list[i]["영어"] == r):
                print("이름: " + self.student_list[i]["이름"] + " 주민번호: " + self.student_list[i]["주민번호"])
         
        print("수학 최고점 : %d" %(s))
        for i in range(0, GradingSystem.ig):
            if(self.student_list[i]["수학"] == s):
                print("이름: " + self.student_list[i]["이름"] + " 주민번호: " + self.student_list[i]["주민번호"])
     
    def findStudent(self, name):
        print("\n---------------추가기능---------------")
        flag = -1
        for x in range(0, GradingSystem.ig):
            if(name == self.student_list[x]["이름"]):
                flag = x
                break
            else : 
                flag = -1
        #찾았다면, 그 학생의 성별과 나이를 계산한다!
        sex = 0
        age = 0
        k = list()
        k = copy.copy(self.student_list[flag]["주민번호"])
        if (flag != -1):
            print("1. " + name + " 의 이름을 가진 학생을 찾았습니다!")
            sex = k[7]
            age = int(k[0:2])
            if(sex == "1" or sex == "3"):
                print(name + "은(는) 남자이고, " , end = "")
                if(sex == "3"):
                    age = age + 100
            else:
                if(sex == "4"):
                    age = age + 100
                print(name + "은(는) 여자이고, " , end = "")
            
            age = 124 - age
            print("%d살 입니다" %(age))
        if (flag != -1):
            
            print("등수: " + str(self.student_list[flag]["등수"]) + " 총점: " + str(self.student_list[flag]["총점"]) + " 주민번호: " + self.student_list[flag]["주민번호"])
        else :
            print("1. 찾지 못했습니다...")
        
    @checktime    
    def rank(self, rank):
        num = 1
        flag = 0
        if(rank == "국어"):
            GradingSystem.temp_list = GradingSystem.language
            tmp = GradingSystem.language[0]["국어"]
        elif(rank == "영어"):
            GradingSystem.temp_list = GradingSystem.english
            tmp = GradingSystem.english[0]["영어"]
        elif(rank == "수학"):
            GradingSystem.temp_list = GradingSystem.math
            tmp = GradingSystem.math[0]["수학"]
        else:
            flag = 1
        
        if(flag == 1):
            print("\n2. 잘못된 입력입니다. ")
            return 
        else:
            print("\n2. 찾으시는 과목의 rank 입니다. ")
            print("%s rank" %(rank))
        
        for i in range(0, GradingSystem.ig):
            if(tmp == GradingSystem.temp_list[i][rank]):
                num = num
            else:
                num += 1
                
            print("등수: " + str(num)+ " 이름: " + GradingSystem.temp_list[i]["이름"] + " 점수: " + str(GradingSystem.temp_list[i][rank]))
            tmp = GradingSystem.temp_list[i][rank]
       
    
    '''
        if(rank == "국어"):
            print("\n2. 찾으시는 과목의 rank 입니다.")
            num = 1
            tmp = GradingSystem.language[0]["국어"]
            print("국어 rank")
            for i in range(0, GradingSystem.ig):
                if(tmp == GradingSystem.language[i]["국어"] ):
                    num = num
                else :
                    num += 1
                print("등수: " + str(num)+ " 이름: " + GradingSystem.language[i]["이름"] + " 점수: " + str(GradingSystem.language[i]["국어"]))
                tmp = GradingSystem.language[i]["국어"]
        
        elif(rank == "영어"):
            print("\n2. 찾으시는 과목의 rank 입니다.")
            num = 1
            tmp = GradingSystem.english[0]["영어"]
            print("영어 rank")
            for i in range(0, GradingSystem.ig):
                if(tmp == GradingSystem.english[i]["영어"]):
                    num = num
                else:
                    num += 1
                print("등수: " + str(num)+ " 이름: " + GradingSystem.english[i]["이름"] + " 점수: " + str(GradingSystem.english[i]["영어"]))
                tmp = GradingSystem.english[i]["영어"]
                
        elif(rank == "수학"):
            print("\n2. 찾으시는 과목의 rank 입니다.")
            num = 1
            tmp = GradingSystem.math[0]["수학"]
            print("수학 rank")
            for i in range(0, GradingSystem.ig):
                if(tmp == GradingSystem.math[i]["수학"]):
                    num = num
                else:
                    num += 1

                print("등수: " + str(num)+ " 이름: " + GradingSystem.math[i]["이름"] + " 점수: " + str(GradingSystem.math[i]["수학"]))
                tmp = GradingSystem.math[i]["수학"]
        else :
            print("\n2. 잘못된 입력입니다!!")
            
        '''
```
